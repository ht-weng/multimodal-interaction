```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating dataset
google_earth_1 = np.array([48.94, 42.09, 68.31, 27.27])
google_earth_2 = np.array([40, 19.43, 29.26, 24.09])
google_map_1 = np.array([20, 12.72, 39.68, 19.90])
google_map_2 = np.array([18.64, 35.54, 12.86, 15.13])
web = np.array([55.26, 30.55, 28.99, 36.62])
data = [google_earth_1, google_earth_2, google_map_1, google_map_2, web]
```

```python
color = ['green', 'red', 'blue', 'purple', 'cyan']
fig = plt.figure(figsize =(15, 10))
 
# Creating plot
bp = plt.boxplot(data)
for box, c in zip(bp['boxes'], color):
    box.set(color=c, linewidth=3)

plt.title("Pilot Study Boxplot", fontsize = 20)
plt.xticks([1, 2, 3, 4, 5], ['GoogleEarth_task1', 'GoogleEarth_task2', 'GoogleMap_task1', 'GoogleMap_task2', 'Web'])
plt.ylabel("Task Completion Time /s", fontsize=16)

# show plot
plt.show()
fig.savefig("pilot_study.pdf", bbox_inches='tight')
```
